```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming you have three numerical columns in your data DataFrame
x = data['from_totally_fake_account']
y = data['monopoly_money_amount']
#z = data['to_randomly_generated_account']  # Assuming 'to_randomly_generated_account' is a numerical column

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y)
ax.set_xlabel('from_totally_fake_account')
ax.set_ylabel('monopoly_money_amount')
#ax.set_zlabel('to_randomly_generated_account')
ax.set_title('3D Scatter Plot')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace this with your actual data)
# Assuming 'from_totally_fake_account' and 'monopoly_money_amount' are your data columns
x_curve = np.linspace(0, 1, 100)
y_curve = np.sin(x_curve * 2 * np.pi) / 2 + 0.5

# Assuming you have two numerical columns in your data DataFrame
df = pd.DataFrame({
    'from_totally_fake_account': np.random.sample(20),
    'monopoly_money_amount': np.random.sample(20)
})

x_scatter = df['from_totally_fake_account']
y_scatter = np.random.sample(20)
z_scatter = df['monopoly_money_amount']

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plot a sin curve using the 'from_totally_fake_account' and 'monopoly_money_amount' columns.
ax.plot(x_curve, y_curve, zs=0, zdir='z', label='curve in (x, y)')

# Plot scatterplot data on the 'from_totally_fake_account' and 'monopoly_money_amount' axes.
ax.scatter(x_scatter, y_scatter, z_scatter, c='r', label='points in (x, z)')

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_xlabel('from_totally_fake_account')
ax.set_ylabel('Y')
ax.set_zlabel('monopoly_money_amount')

# Customize the view angle
ax.view_init(elev=20., azim=-35)

plt.show()
```

```python
# Visualize the distribution of categorical variables using count plots
plt.figure(figsize=(4,5))
sns.countplot(y='to_randomly_generated_account', data=data, order=data['to_randomly_generated_account'].value_counts().index[:10])
plt.title('Top 10 most frequent to_randomly_generated_account')
plt.xlabel('Count')
plt.ylabel('to_randomly_generated_account')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace this with your actual data)
categories = data['to_randomly_generated_account'].value_counts().index[:10]
counts = data['to_randomly_generated_account'].value_counts().values[:10]

# Create a figure and a polar axis
fig, ax = plt.subplots(figsize=(5,6), subplot_kw={'projection': 'polar'})

# Calculate the angle for each category
theta = np.linspace(0, 2 * np.pi, len(categories), endpoint=False)

# Normalize counts to use as heights for bars
normalized_counts = counts / np.max(counts)

# Map normalized counts to colors using a colormap
cmap = plt.get_cmap('viridis')
colors = cmap(normalized_counts)

# Plot the bars
bars = ax.bar(theta, normalized_counts, width=0.5, alpha=0.7, color=colors)

# Set the labels for each bar
ax.set_xticks(theta)
ax.set_xticklabels(categories, fontsize=8)

# Add title and adjust layout
plt.title('Top 10 most frequent to_randomly_generated_account', fontsize=5, pad=10)
plt.tight_layout()

plt.show()
```

### Float to int conversion 

```python
# Extracting two columns from the dataset
selected_columns = data[['from_totally_fake_account', 'monopoly_money_amount']]
selected_columns
```

```python
# Convert 'from_totally_fake_account' column to integer
data['from_totally_fake_account'] = data['from_totally_fake_account'].astype(int)

# Round 'monopoly_money_amount' column to the nearest integer
data['monopoly_money_amount'] = data['monopoly_money_amount'].round().astype(int)
```
